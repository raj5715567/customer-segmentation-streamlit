```python
!pip install pandas
import pandas as pd

df = pd.read_csv("data/Customer Personality Analysis Dataset.csv")
df.head()
```

```python
df.isnull()
```

```python
df.describe()
```

```python
# Remove missing values
df = df.dropna()

# Check again
df.info()
```

```python
df['Year_Birth'] = pd.to_numeric(df['Year_Birth'], errors='coerce')
```

```python
df['Age'] = 2024 - df['Year_Birth']
```

```python
df['Total_Spending'] = (
    df['MntWines'] +
    df['MntFruits'] +
    df['MntMeatProducts'] +
    df['MntFishProducts'] +
    df['MntSweetProducts'] +
    df['MntGoldProds']
)
```

```python
df[['Total_Spending']].head()
```

```python
df['CLV'] = df['Total_Spending'] * df['NumWebPurchases']
```

```python
df[['CLV']].head()
```

```python
df['Income'] = pd.to_numeric(df['Income'], errors='coerce')
df = df.dropna(subset=['Income'])
```

```python
X = df[['Income', 'Age', 'Total_Spending']]
```

```python
!pip install scikit-learn
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
!pip install matplotlib
import matplotlib.pyplot as plt

plt.scatter(df['Income'], df['Total_Spending'], c=df['Cluster'])
plt.xlabel("Income")
plt.ylabel("Total Spending")
plt.title("Customer Segments")
plt.show()
```

```python
df.groupby('Cluster')[['Income', 'Age', 'Total_Spending']].mean()
```

```python
df = df[df['Income'] < 200000]
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Income'], df['Total_Spending'], c=df['Cluster'])
plt.xlabel("Income")
plt.ylabel("Total Spending")
plt.title("Customer Segmentation")
plt.show()
```
